أنشئ دالة لاستخراج المهارات من أي نص

```python
## دالة لاستخراج المهارات الموجودة في السيرة الذاتية
def extract_skills(text, skills_list):
    text = text.lower()

    found_skills = []

    for skill in skills_list:
        if skill.lower() in text:
            found_skills.append(skill)

    return sorted(list(set(found_skills)))
```

```python
# استخراج المهارات من السيرة الذاتية وعرضها
found_skills = extract_skills(resume, skills)

print("Extracted Skills:")
print(found_skills)
```
