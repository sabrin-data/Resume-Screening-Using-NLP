```python
#تحميل الداتا النظيفة
import pandas as pd

df = pd.read_csv("resume_clean_dataset.csv")

df.head()
```

```python
#فحص حجم الداتا
print("Dataset shape:", df.shape)
print("Number of categories:", df["category"].nunique())
```

```python
#توزيع التصنيفات
df["category"].value_counts()
```

```python
#رسم عدد السير بكل تصنيف
import matplotlib.pyplot as plt

category_counts = df["category"].value_counts()

plt.figure(figsize=(12, 6))
category_counts.plot(kind="bar")
plt.title("Number of Resumes per Category")
plt.xlabel("Category")
plt.ylabel("Number of Resumes")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#تحليل طول النصوص
df["clean_text_length"].describe()
```

```python
#رسم توزيع أطوال السير
plt.figure(figsize=(10, 5))
plt.hist(df["clean_text_length"], bins=30)
plt.title("Distribution of Resume Text Lengths")
plt.xlabel("Text Length")
plt.ylabel("Number of Resumes")
plt.tight_layout()
plt.show()
```
